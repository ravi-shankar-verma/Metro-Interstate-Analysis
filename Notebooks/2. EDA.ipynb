#Further Exploration

##Traffic Volume By Hour

<summary><font size="3" color="#ADD8E6"><b>Speculation</b></font></summary>

*   Traffic volume and hour are the most correlated features in this dataframe.
* I will observe this by day of week


```python
hr_tr = df.groupby('hour', as_index=False)['traffic_volume'].mean()

plt.figure(figsize=(15, 7))
barplot = sns.barplot(data=hr_tr, x='hour', y='traffic_volume')

min_val, max_val = hr_tr['traffic_volume'].min(), hr_tr['traffic_volume'].max()
min_pos, max_pos = hr_tr['traffic_volume'].idxmin(), hr_tr['traffic_volume'].idxmax()

plt.axhline(min_val, color='green', linestyle='--', linewidth=2, label=f'Min: {min_val:.0f}')
plt.axhline(max_val, color='red', linestyle='--', linewidth=2, label=f'Max: {max_val:.0f}')

barplot.patches[min_pos].set_color('green')
barplot.patches[max_pos].set_color('red')

plt.title('\nMean Traffic Volume by Hour\n')
plt.legend()
plt.show()
```

```python
list_days = df['day_of_week'].unique().tolist()
```

```python
for day in list_days:
    day_hr = df.loc[df['day_of_week'] == day].groupby('hour', as_index=False)['traffic_volume'].mean()

    plt.figure(figsize=(10, 5))
    barplot = sns.barplot(data=day_hr, x='hour', y='traffic_volume')

    min_pos, max_pos = day_hr['traffic_volume'].idxmin(), day_hr['traffic_volume'].idxmax()
    min_val, max_val = day_hr.loc[min_pos, 'traffic_volume'], day_hr.loc[max_pos, 'traffic_volume']

    for val, pos, color in zip([min_val, max_val], [min_pos, max_pos], ['green', 'red']):
        plt.axhline(val, color=color, linestyle='--', linewidth=2, label=f'{color.capitalize()}: {val:.0f}')
        barplot.patches[pos].set_color(color)

    plt.title(f'\nMean Traffic Volume by Hour ({day})\n')
    plt.legend()
    plt.show()
```

<summary><font size="3" color="#ADD8E6"><b>Findings</b></font></summary>

* Weekends represent a significant distribution change from weekdays.
* It may be useful to one hot encode the day of week for modeling performance.

##Traffic Volume by Temperature

<summary><font size="3" color="#ADD8E6"><b>Speculation</b></font></summary>

*   Traffic volume and temperature have a weak correlation.
* I will observe this by hour.


```python
mean_data = df.groupby(df['hour']).agg({'temp': 'mean', 'traffic_volume': 'mean'})

fig, ax1 = plt.subplots(figsize=(10, 5))

color = 'tab:blue'
ax1.set_xlabel('Hour')
ax1.set_ylabel('Temperature', color=color)
ax1.plot(mean_data.index, mean_data['temp'], color=color, label='Temperature')
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()
color = 'tab:red'
ax2.set_ylabel('Traffic Volume', color=color)
ax2.plot(mean_data.index, mean_data['traffic_volume'], color=color, label='Traffic Volume', lw=3)
ax2.tick_params(axis='y', labelcolor=color)

lines, labels = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax1.legend(lines + lines2, labels + labels2, loc='upper left')

plt.title('\nMean Temperature and Traffic Volume by Hour\n')
```

```python
for day in list_days:
    day_data = df.loc[df['day_of_week'] == day].groupby('hour', as_index=False).agg({'temp': 'mean', 'traffic_volume': 'mean'})

    fig, ax1 = plt.subplots(figsize=(10, 5))

    ax1.set_xlabel('Hour')
    ax1.set_ylabel('Temperature', color='tab:blue')
    ax1.plot(day_data['hour'], day_data['temp'], color='tab:blue', label='Temperature')
    ax1.tick_params(axis='y', labelcolor='tab:blue')

    ax2 = ax1.twinx()
    ax2.set_ylabel('Traffic Volume', color='tab:red')
    ax2.plot(day_data['hour'], day_data['traffic_volume'], color='tab:red', label='Traffic Volume', lw=3)
    ax2.tick_params(axis='y', labelcolor='tab:red')

    lines, labels = ax1.get_legend_handles_labels()
    lines2, labels2 = ax2.get_legend_handles_labels()
    ax1.legend(lines + lines2, labels + labels2, loc='upper left')

    plt.title(f'\nMean Temperature and Traffic Volume by Hour ({day})\n')
    plt.show()
```

<summary><font size="3" color="#ADD8E6"><b>Findings</b></font></summary>

*   Traffic volume and temperature reasonably well correlated later in the day, not so well in the morning.
* Could be worth binning temperature to improve model performance.
* I could potentially observe this by season to improve performance further.


##Traffic Volume by Weather Condition

<summary><font size="3" color="#ADD8E6"><b>Speculation</b></font></summary>

*  I test a personal theory that rain will slow down traffic rate.
* I will test other weather conditions alongside.

```python
weather_list = df['weather_description'].unique()

clear_weather = df.loc[df['weather_description'] == 'clear'].groupby('hour').agg({'traffic_volume':'mean'})
all_other_weather = df.loc[df['weather_description'] != 'clear'].groupby('hour').agg({'traffic_volume':'mean'})

for weather in weather_list:
  if weather != 'clear':
    specific_weather = df.loc[df['weather_description'] == weather].groupby('hour').agg({'traffic_volume':'mean'})
    fig, ax1 = plt.subplots(figsize=(10,5))
    sns.barplot(data = specific_weather,
                x = specific_weather.index,
                y = 'traffic_volume', ax = ax1)
    ax1.set_xlabel('Hour')
    ax1.set_ylabel('Traffic Volume', color='blue')
    ax1.tick_params(axis='y', labelcolor='blue')
    ax2 = ax1.twinx()
    sns.lineplot(data = clear_weather,
                x = clear_weather.index,
                y = 'traffic_volume',
                color = 'red',
                ax = ax2)
    ax2.set_ylabel('Traffic Volume (Clear)', color='red')
    ax2.tick_params(axis='y', labelcolor='red')
    lines1, labels1 = ax1.get_legend_handles_labels()
    lines2, labels2 = ax2.get_legend_handles_labels()

    plt.title(f'\n Mean Traffic Volume for {weather} weather\n')
    plt.show()
```

<summary><font size="3" color="#ADD8E6"><b>Findings</b></font></summary>

* Very little change between distributions of traffic volume by weather condition
* My personal hypothesis of rain slowing traffic down not well observed.


##Traffic Volume by Holiday

<summary><font size="3" color="#ADD8E6"><b>Speculation</b></font></summary>

*   I would expect to see changes in the distribution of traffic depending on holiday

```python
plt.figure(figsize=(13,6))
sns.barplot(
    data = df,
    x = 'holiday',
    y = 'traffic_volume',
    ci = None
)

plt.xticks(rotation=75)
plt.title('\nMean Traffic Volume by Holiday Status\n')
```

```python
list_holidays = df['holiday'].unique().tolist()
```

```python
for holiday in list_holidays:
    holiday_hr = df.loc[df['holiday'] == holiday].groupby('hour', as_index=False)['traffic_volume'].mean()

    plt.figure(figsize=(10, 5))
    barplot = sns.barplot(data=holiday_hr, x='hour', y='traffic_volume')

    min_pos, max_pos = holiday_hr['traffic_volume'].idxmin(), holiday_hr['traffic_volume'].idxmax()
    min_val, max_val = holiday_hr.loc[min_pos, 'traffic_volume'], holiday_hr.loc[max_pos, 'traffic_volume']

    for val, pos, color in zip([min_val, max_val], [min_pos, max_pos], ['green', 'red']):
        plt.axhline(val, color=color, linestyle='--', linewidth=2, label=f'{color.capitalize()}: {val:.0f}')
        barplot.patches[pos].set_color(color)

    plt.title(f'\nMean Traffic Volume by Hour ({holiday})\n')
    plt.legend()
    plt.show()
```

<summary><font size="3" color="#ADD8E6"><b>Findings</b></font></summary>

* Reasonable change in distribution and timings of traffic depending on holiday.
* One Hot Encoding will present good opportunity to improve model performance.
